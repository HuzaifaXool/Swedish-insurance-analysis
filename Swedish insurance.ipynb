```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
DF=pd.read_csv('swedish_insurance.csv')
```

```python
DF.head(10)
```

```python
plt.figure(figsize=(12,5))
plt.title('Box Plot for Data')
sns.boxplot(DF)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(DF)
plt.title('Scatter Plot for dataset')
plt.show()
```

```python
mean_x=DF['X'].mean()
mean_y=DF['Y'].mean()
variance_x=DF['X'].var()
variance_y=DF['Y'].var()
```

```python
print(f'The mean of number of claims (X) is {mean_x}\nThe mean of total payment (Y) is {mean_y}\nThe Variance of number of claims (X) is {variance_x}\nThe Variance of number of claims (X) is {variance_y}')
```

```python
covariance=DF['X'].cov(DF['Y'])
```

```python
print(covariance)
```

```python
b1=covariance/variance_x
b0=mean_y-b1*mean_x
```

```python
print("Coefficient b1 (slope):", b1)
print("Coefficient b0 (intercept):", b0)
```

```python
DF['Y_hat']=b0+b1*DF['X']
```

```python
DF.head()
```

```python
DF['Residuals']=DF['Y']-DF['Y_hat']
```

```python
DF.head()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(DF['X'], DF['Y'], color='blue', label='Original Data', marker='o')
plt.plot(DF['X'], DF['Y_hat'], color='red', label='Regression Line',marker='*')
plt.title('Linear Regression Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```
